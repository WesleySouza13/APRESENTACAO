#definição x e y

```python
y = df['quality']
x = df.drop(['quality','Id'], axis = 1)
```

```python
y
```

#importando modelo de treino

```python
from sklearn.model_selection import train_test_split
```

```python
#criando variavel teste e treino
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size = 0.3)
```

criamos a variavel para teste e treino a partir do modelo train_test_split

#criando modelo de previsao

```python
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

#imprimindo resultados
resultados = modelo.score(x_teste, y_teste)
print('acuracia:', resultados)
```
